```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0,2,100)

y = x*x

z  = x**3

#Plot in matlpotlib

plt.plot(x,x,label="X") 
plt.plot(x,y ,label="y")
plt.plot(x,z,label="z")
plt.title("This is matplotlib")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend()

#as comparativel matplotlib has more controlle than seaborn

# but we want more shiny kind plots and graphs we can used plotly.express library plots


```

```python
x_ba = np.arange(0,10,0.1)

y_ba = np.sin(x_ba)

z_ba = np.cos(y_ba)

plt.plot(x_ba,y_ba,label="x",color="green",marker="*")
plt.plot(x_ba,z_ba,label="y",color="yellow",marker=".")
plt.title("x sin y cos")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```
