```python
# Grouing or aggregation in pandas


train.shape

# count also provide us the non-null value counts
# train.count()

# Mean = Average
# Median = Central Value
# Mode = Repeated Value


train.Age.median()

train.Age.max()

# group by

train.groupby("Sex")['Fare'].sum()

train.groupby("Sex")['Survived'].count()

#single value group by

train.groupby("Sex")[['Survived','Pclass']].count()


Test = train.groupby("Sex")

#Get_group by help us to find the value wihin the group

Test.get_group('male')['Survived'].mean()

#Multiple group by
train.groupby(["Sex","Pclass"])['Fare'].sum()

train.groupby(["Sex","Pclass"])[['Fare','Survived']].sum()

#Group by with agg function

# train.groupby(["Sex","Pclass"]).agg({'Fare':"sum","Age":"median"})

```

```python
#Cross Tab

#Cross tab is very handy technique if we are dealing with categorical values
#we can cross tabulation using the cross tab
#cross tab are very handy to get some quick insight from data

pd.crosstab(train.Sex,train.Pclass)

#Cross tab with multiple values


pd.crosstab([train.Sex,train.Survived],train.Pclass)

#Margin = Total

#https://pbpython.com/pandas-crosstab.html ***** resources


pd.crosstab(train.Sex,train.Pclass,margins=True)

pd.crosstab(train.Sex,train.Pclass).plot(kind="bar")

#Stack is just like unpivot in pandas

train[['Sex','Pclass']].stack()
```

```python
#Pivot table

#Single value pivot

train.pivot_table(index="Sex",columns="Pclass",values="Fare",aggfunc = "sum")

#Multiple Value pivot

train.pivot_table(index=["Sex","Survived"],columns="Pclass",values="Fare",aggfunc = "sum")

#train.groupby(["Sex","Pclass"]).Fare.sum()

#Unstack perform same operation with group by like pivot

#train.groupby(["Sex","Pclass"]).Fare.sum().unstack()


```
