```python
import pandas as pd

import seaborn as sns

train_path = "/content/train.csv"

df = pd.read_csv(train_path)

# df['Age'].hist(bins=10)


# df['Age'].plot(kind="hist",title='My title',xlabel="Age")

df['Age'].plot(kind="box",rot=45,title="Age Box Plot",color="c")

# df.plot.scatter(x="Fare",y="Age",alpha=1)



# df['Sex'].value_counts().plot(kind="bar")
```

```python
df.head(10)

# sns.boxplot(x="Pclass",y="Fare",hue="Sex",data=df)

# sns.boxplot(x="Sex",y="Survived",data=df)

df = sns.load_dataset("tips")
df.columns

# tips
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df)
```

```python
# time/tips

# tips.columns


sns.boxplot(x="time",y="tip",hue="day",data=df)
```

```python
#day/tips/smoker

sns.boxplot(x="day",y="tip",hue="smoker",palette="Set3",data=df)
```

```python
# SMOKER WISE TOTAL BILLS

sns.boxplot(x="smoker",y="total_bill",hue="day",palette="Set3",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue="time",data=df,linewidth=2.5)
```

```python
sns.boxplot(x="time",y="tip",data=df,order=["Dinner","Lunch"])
```

```python
sns.boxplot(orient="h",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df,order=['Sat','Sun','Thur','Fri'])

# #Attribute for boxplot
# x axis,
# y axis,
# data ,
# order ,
# orient ,
# palette,
# linewidth

```

```python
#Violin plot in seaborn

sns.violinplot(x="time",y="tip",data=df)
```

```python
###histplot

sns.histplot(data=df,x="total_bill")
```

```python
# Bar Plot

# sns.barplot(data=df,x="day",y="total_bill")


sns.barplot(data=df,x="day",y="total_bill",hue="time",palette="Set3")
```

```python
#Scatter plot required two numeric values 

# sns.scatterplot(data=df,x="tip",y="total_bill")

sns.scatterplot(data=df,x="total_bill",y="tip")
```
